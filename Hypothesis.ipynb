```python
import numpy as np
import scipy.stats as stats
```

```python
x_bar = 3050        # sample mean
mu = 4000           # theoretical mean
sigma = 125         # population standard deviation
n = 25              # sample size
```

```python
z = (x_bar - mu) / (sigma / np.sqrt(n))
z
```

```python
p_value = 1 - stats.norm.cdf(z)
p_value
```

```python
z_critical = stats.norm.ppf(1 - 0.05)
z_critical
```
