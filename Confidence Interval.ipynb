```python
import numpy as np
import scipy.stats as stats
```

```python
data = np.array([
    1.13, 1.55, 1.43, 0.92, 1.25,
    1.36, 1.32, 0.85, 1.07, 1.48,
    1.20, 1.33, 1.18, 1.22, 1.29
])
```

```python
n = len(data)                
mean = np.mean(data)         
std_sample = np.std(data, ddof=1)  
mean, std_sample, n
```

```python
alpha = 0.01
df = n - 1

t_critical = stats.t.ppf(1 - alpha/2, df)
t_critical
```

```python
margin_error_t = t_critical * (std_sample / np.sqrt(n))
margin_error_t
```

```python
ci_lower_t = mean - margin_error_t
ci_upper_t = mean + margin_error_t

ci_lower_t, ci_upper_t
```
